```python
import re
import pandas as pd

def find_and_extract_links(dataframe, column_name, new_column_name='ExtractedLink'):
    # Define a regular expression pattern to match URLs
    url_pattern = r'https?://\S+|www\.\S+'

    # Lists to store extracted links
    extracted_links = []

    # Iterate through the rows of the DataFrame
    for index, row in dataframe.iterrows():
        text = str(row[column_name])  # Convert to string in case the column contains non-string types

        # Find all matches of the pattern in the text
        links = re.findall(url_pattern, text)

        # Use the first link found or 'N/A' if no link is present
        extracted_link = links[0] if links else 'N/A'

        # Append the extracted link to the list
        extracted_links.append(extracted_link)

    # Add a new column to the DataFrame with the extracted links
    dataframe[new_column_name] = extracted_links

    # Save the updated DataFrame back to the CSV file
    dataframe.to_csv('updated_data.csv', index=False)


# Establish a local path to the csv file and read it in to a dataframe
local_path = '/tweets_data_w_timestamp.csv'
df = pd.read_csv(local_path)

# Apply the extract_links function to the dataframe
find_and_extract_links(df, 'Text', new_column_name='ExtractedLink')
```
