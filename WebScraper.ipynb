```python
from ntscraper import Nitter

scraper = Nitter(log_level=1, skip_instance_check=False)
```

```python
tweets = scraper.get_tweets("@econ_ra", mode='term', until='2022-05-12', exclude=['replies'])
```

```python
print(tweets)
```

```python
# Extract the 'tweets' list from the scraper result
tweets_str_list = tweets.get('tweets', [])

print(tweets_str_list)
```

```python
import os
import json

def save_tweets_to_txt(tweets):
    # Create a directory to store the text files if it doesn't exist
    if not os.path.exists('tweet_texts'):
        os.makedirs('tweet_texts')

    # Iterate through each tweet and save it to a separate text file
    for i, tweet in enumerate(tweets, start=1):
        # Convert the tweet dictionary to a string
        tweet_str = json.dumps(tweet)

        with open(f'tweet_texts/tweet_{i}.txt', 'w', encoding='utf-8') as file:
            file.write(tweet_str)
            print(f"Tweet {i} saved to 'tweet_texts/tweet_{i}.txt'")

# Example usage:
# Assuming 'tweets_str_list' is a list of dictionaries containing tweet information
save_tweets_to_txt(tweets_str_list)
```

```python
import csv

# Create csv file instance
csv_filename = 'tweets_data_w_timestamp3.csv'

# Define the header for the CSV file
csv_header = ['Link', 'Text', 'User Name', 'Timestamp']

# Open the CSV file in write mode
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    # Create a CSV writer object
    csv_writer = csv.writer(csv_file)

    # Write the header to the CSV file
    csv_writer.writerow(csv_header)

    # Write each tweet's information to the CSV file (in order of the header)
    for tweet in tweets_str_list:
        row_data = [
            tweet.get('link', ''),
            tweet.get('text', ''),
            tweet['user'].get('username', ''),
            tweet.get('date', ''),
        ]
        csv_writer.writerow(row_data)

print(f"CSV file '{csv_filename}' has been created.")
```
